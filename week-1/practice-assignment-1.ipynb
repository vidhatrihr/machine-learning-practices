```python
def convert_to_seconds(duration):
  seconds = 0
  if ' ' in duration:
    hrs, mins = duration.split()
    seconds += int(hrs[:-1]) * 3600
    seconds += int(mins[:-1]) * 60

  elif duration[-1] == 'h':
    seconds += int(duration[:-1]) * 3600

  elif duration[-1] == 'm':
    seconds += int(duration[:-1]) * 60

  return seconds


convert_to_seconds('3h 3m')
convert_to_seconds('3h')
convert_to_seconds('3m')
```

```python
df['Duration'].map(convert_to_seconds).mean()
```

```python
def time_to_category(time):
  hrs = int(time[:2])

  if 5 <= hrs < 12:
    return 'morning'
  elif 12 <= hrs < 17:
    return 'afternoon'
  elif 17 <= hrs < 20:
    return 'evening'
  else:
    return 'night'


time_to_category('04:00')
```

```python
mask1 = df['Dep_Time'].map(time_to_category) == 'morning'
mask2 = df['Arrival_Time'].map(time_to_category) == 'evening'
(mask1 & mask2).sum()
```
