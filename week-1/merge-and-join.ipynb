```python
import pandas as pd
```

```python
df1 = pd.DataFrame({

    "ID": [10001, 10002, 30003, 40004, 50005, 9001],
    "Qty": [10, 20, 20, 40, 50, 20],
    "Category": ["A", "B", "C", "D", "E", 'E'],
})

df2 = pd.DataFrame({
    "ID": [10001, 10002, 30003, 60006, 70007, 9004],
    "Qty": [10, 20, 30, 40, 60, 20],
    "City": ["Lucknow", "HBH", "Chennai", "Delhi", "Jaipur", 'HBH']
})
```

```python
df1.merge(df2)  # ID and Qty of df1 and City of df2
```

```python
df1.merge(df2, how='inner')
```

```python
df1.merge(df2, how='inner', on='ID')
df2.merge(df1, how='inner', on='ID')
```

```python
df1.merge(df2, how='inner', on='Qty')
```

```python
df1.merge(df2, how='inner', on=['ID', 'Qty'])
```

```python
df1.merge(df2, how='outer')
```

```python
df1.merge(df2, how='outer', on='ID')
```

```python
df1.merge(df2, how='left')
```

```python
df1.merge(df2, how='right')
df1.merge(df2, how='left')
```

```python
df1.merge(df2, how='cross')
```

```python
# Merge two tables so you only get employees whose departments exist.

employees = pd.DataFrame({
    "emp_id": [101, 102, 103, 104],
    "name": ["Alice", "Bob", "Charlie", "David"],
    "dept_id": [10, 20, 30, 40]
})

departments = pd.DataFrame({
    "dept_id": [10, 20, 30],
    "dept_name": ["HR", "Finance", "IT"]
})

employees.merge(departments, on='dept_id')
```

```python
# Some students mistakenly got registered for courses that donâ€™t exist.
# Merge so that all students appear, regardless of whether their course exists.

students = pd.DataFrame({
    "roll_no": [1, 2, 3, 4],
    "student_name": ["John", "Emma", "Liam", "Sophia"],
    "course_id": [101, 102, 103, 104]
})

courses = pd.DataFrame({
    "course_id": [101, 102, 105],
    "course_name": ["Math", "Physics", "Biology"]
})

students.merge(courses, how='left')
```

```python
# An e-commerce store has sales records for two years. The manager wants to check which orders belong to:
# 1. only 2024,
# 2. only 2025,
# 3. both years.

sales_2024 = pd.DataFrame({
    "order_id": [1, 2, 3],
    "product": ["Laptop", "Phone", "Tablet"]
})

sales_2025 = pd.DataFrame({
    "order_id": [3, 4, 5],
    "product": ["Tablet", "Monitor", "Headphones"]
})

mapping = {
    'left_only': 'Only 2024',
    'right_only': 'Only 2025',
    'both': 'Both years',
}

merged_df = pd.merge(sales_2024, sales_2025, on=[
                     'order_id', 'product'], how='outer', indicator=True)

merged_df['status'] = merged_df['_merge'].map(mapping)

sales_2024.merge(sales_2025, how='left')
sales_2024.merge(sales_2025, how='right')
```
