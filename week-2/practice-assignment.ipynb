```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/DataPreprocessingGraded_dataset.csv')
df.head(10)
```

```python
df.isnull().sum().sum()
(df.values == '?').sum()
```

```python
df = df.replace('?', np.nan)
df.isna().sum()
```

```python
df = df.astype({'V1': float, 'V2': float})
df.info()
```

```python
df.describe()
```

```python
df['Target'].value_counts()
df['V5'].value_counts()
```

```python
numeric_cols = df[['V1', 'V2', 'V3', 'V4']]

outlier_count = {}

for col in numeric_cols.columns:
  col_data = df[col].dropna()

  Q1 = col_data.quantile(0.25)
  Q3 = col_data.quantile(0.75)

  iqr = Q3 - Q1

  lower_bound = Q1 - 1.5 * iqr
  upper_bound = Q1 + 1.5 * iqr

  outlier = col_data[(col_data < lower_bound) | (col_data > upper_bound)]
  outlier_count[col] = len(outlier)
  print(f'outlier count of {col} = {outlier_count[col]}')
```
