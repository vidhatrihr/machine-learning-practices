```python
import numpy as np
```

```python
X = np.array([
    [2, 2],
    [4, 2],
    [-1, 2],
    [5, 2],
    [8, 2],
])  # B x M

X.shape
```

- B: # of batch
- M: # of feature
- N: # of neurons

```python
W = np.array([[2, 1], [2, 2]], dtype=float)  # a 2D arr shape M x N
b = np.array([0, 1])  # arr of shape N

y = X @ W + b
y
```

```python
lr = 0.01

W -= lr * X.sum(axis=0).reshape(-1, 1)
y = X @ W + b
y
```
