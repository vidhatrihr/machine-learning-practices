## The Normalizer

```python
np.random.seed(1)
data = np.random.randint(0, 100, (10, 4))
data
```

```python
_min = data.min(axis=0)
_max = data.max(axis=0)
(data - _min) / (_max - _min)
```

```python
data = np.array([[37, 12, 72,  9],
                [75,  5, 79, 64],
                [1,  0, 60, 81]])

_min = data.min(axis=0)
_max = data.max(axis=0)

transformed_data = (data - _min) / (_max - _min)
transformed_data
```

## The Outlier Finder

```python
np.random.seed(1)

data = np.random.randn(100)

# Mean before replacement
data.mean()

twice_std = 2 * data.std()
dist = np.abs(data - data.mean())
mask = dist > twice_std
data[mask] = 99.9

# Mean after replacement
data.mean()
```

```python
data = np.array([2, 3, 4, 5])
data.mean()
data.std()

mask = np.abs(data - data.mean()) < data.std()
data[mask]
```

## Image-Style Manipulations (Channel Swapping)

```python
np.random.seed(1)
data = np.random.randint(0, 256, (10, 10, 3))

data[:, :, 1] = 0  # "Mute" the Green channel
data.mean()  # brightness

red = data[:, :, 0]
blue = data[:, :, 2]
mask = (red > 200) & (blue < 50)
mask.sum()
```

```python
np.random.seed(1)

data = np.random.randint(0, 10, (2, 2, 3))
data[:, :, 1] = 0  # "Mute" the Green channel
data
data[:, :, 0] > 5  # red channel
data[:, :, 2] > 5  # blue channel

data.mean()  # brightness
```
