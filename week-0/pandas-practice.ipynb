```python
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_population_(United_Nations)'

# Pass the 'User-Agent' header to mimic a browser
tables = pd.read_html(
    url,
    storage_options={'User-Agent': 'Mozilla/5.0'}
)

df_population = tables[0]
df_population
```

```python
df_population['UN continental region[1]'].value_counts()
df_population.describe()
```

```python
mask = df_population['Population (1 July 2022)'] < 10000
mask.sum()
df_population[mask]
```

```python
mask = (df_population['Population (1 July 2022)'] < 10000) & (
    df_population['UN continental region[1]'] == 'Oceania')

mask.sum()
df_population[mask]
```

```python
my_data = {
    'name': [None, 'vishnu', 'paru', 'rahul', 'maa'],
    'age': [None, 30, None, 59, 70],
    'marks': [90, None, 30, None, 100],
}

df = pd.DataFrame(my_data)
df
```

```python
df.info()
df.isnull()
df.isnull().sum()
```

```python
df
df.dropna()
```

```python
df.fillna(0)
df['name'].fillna('Unknown')
df[['age', 'marks']].fillna(0)
df.fillna({
    'age': 20,
    'marks': 30,
})
```

```python
df[['age', 'marks']].fillna({
    'age': df['age'].mean(),
    'marks': df['marks'].mean()
}, inplace=True)
```

```python
df['salary'] = df['age'] * df['marks']

df
```

```python
df.columns
df.dtypes
df.select_dtypes(include=['number'])  # numeric cols
df.select_dtypes(include=['object'])  # categorical cols
```
