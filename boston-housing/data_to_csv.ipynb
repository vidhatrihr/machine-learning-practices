```python
import re
```

```python
with open('data/housing.data') as file:
  data = file.read()
```

```python
col_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE',
             'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']

header = ','.join(col_names)
```

```python
content = header + '\n'
for line in data.splitlines():
  content += re.sub(r'\s+', ',', line.strip()) + '\n'
```

```python
with open('data/housing.csv', 'w') as file:
  file.write(content)
```
